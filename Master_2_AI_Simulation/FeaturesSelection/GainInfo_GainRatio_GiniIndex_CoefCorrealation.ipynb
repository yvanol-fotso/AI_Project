```python
import numpy as np
from sklearn.preprocessing import LabelEncoder

def calculate_pearson_correlation(data):
    features = np.array(data)[:, :-1]  # Séparer les caractéristiques
    labels = np.array(data)[:, -1]     # Séparer les étiquettes de classe

    # Convertir les étiquettes de classe en valeurs numériques
    label_encoder = LabelEncoder()
    labels_encoded = label_encoder.fit_transform(labels)

    # Calculer le coefficient de corrélation de Pearson entre chaque caractéristique et la classe
    pearson_correlation = [np.abs(np.corrcoef(feature.astype(float), labels_encoded)[0, 1]) for feature in features.T]

    return pearson_correlation

# Exemple d'utilisation
data = [
    [1, 2, 3, 'Attack'],
    [2, 3, 4, 'Normal'],
    [3, 4, 5, 'Attack'],
    [4, 5, 6, 'Normal'],
    [5, 6, 7, 'Attack'],
    [6, 7, 8, 'Normal'],
]

pearson_correlation = calculate_pearson_correlation(data)

# Afficher les coefficients de corrélation de Pearson pour chaque caractéristique
for i, correlation in enumerate(pearson_correlation):
    print(f"Feature {i+1}: Pearson Correlation = {correlation}")
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder

def calculate_pearson_correlation(data):
    features = np.array(data)[:, :-1]  # Séparer les caractéristiques
    labels = np.array(data)[:, -1]     # Séparer les étiquettes de classe

    # Encoder les étiquettes de classe en valeurs numériques
    label_encoder = LabelEncoder()
    encoded_labels = label_encoder.fit_transform(labels)

    # Calculer le coefficient de corrélation de Pearson entre chaque caractéristique et les étiquettes encodées
    pearson_correlation = [np.abs(np.corrcoef(feature.astype(float), encoded_labels.astype(float))[0, 1]) for feature in features.T]

    return pearson_correlation

# Exemple d'utilisation
data = [
    [1, 2, 3, 'Attack'],
    [2, 3, 4, 'Normal'],
    [3, 4, 5, 'Attack'],
    [4, 5, 6, 'Normal'],
    [5, 6, 7, 'Attack'],
    [6, 7, 8, 'Normal'],
]

pearson_correlation = calculate_pearson_correlation(data)

# Afficher les coefficients de corrélation de Pearson pour chaque caractéristique
for i, correlation in enumerate(pearson_correlation):
    print(f"Feature {i+1}: Pearson Correlation = {correlation}")
```
