```python
import pandas as pd
import numpy as np

# Chargement des fichiers CSV
benin_data = pd.read_csv("../../data/Bell_DNS Dataset/features_domain_benign_csv.csv")
malware_data = pd.read_csv("../../data/Bell_DNS Dataset/features-domain_Malware.csv")
phishing_data = pd.read_csv("../../data/Bell_DNS Dataset/features-domain_phishing.csv")
spam_data = pd.read_csv("../../data/Bell_DNS Dataset/features-domain_Spam.csv")

# Ajout de la colonne 'Class'
benin_data['Class'] = 'Benin'
malware_data['Class'] = 'Malware'
phishing_data['Class'] = 'Phishing'
spam_data['Class'] = 'Spam'

# Concaténation des données de toutes les classes
all_data = pd.concat([benin_data, malware_data, phishing_data, spam_data], ignore_index=True)

# Comptage des échantillons par classe
class_counts = all_data['Class'].value_counts()

# Détermination du nombre minimum d'échantillons dans une classe
min_samples = class_counts.min()

# Équilibrage des classes en sélectionnant un sous-ensemble aléatoire pour chaque classe
balanced_data = pd.DataFrame()
for class_name in class_counts.index:
    class_subset = all_data[all_data['Class'] == class_name].sample(min_samples, random_state=42)
    balanced_data = pd.concat([balanced_data, class_subset], ignore_index=True)

# Affichage des nouvelles statistiques des classes équilibrées
print(balanced_data['Class'].value_counts())
print(balanced_data.shape)

# Extraction de la variable cible (y)
y = balanced_data['Class']
```

```python
from sklearn.utils import resample

# Déterminer la taille moyenne des jeux de données
average_size = sum([len(dataset) for dataset in [benin_data, malware_data, phishing_data, spam_data]]) // 4

# Rééquilibrer chaque jeu de données
benin_resampled = resample(benin_data, replace=True, n_samples=average_size, random_state=42)
malware_resampled = resample(malware_data, replace=True, n_samples=average_size, random_state=42)
phishing_resampled = resample(phishing_data, replace=True, n_samples=average_size, random_state=42)
spam_resampled = resample(spam_data, replace=True, n_samples=average_size, random_state=42)

# Afficher la taille des ensembles de données rééquilibrés
print("Taille de l'ensemble de données rééquilibré pour Benin:", len(benin_resampled))
print("Taille de l'ensemble de données rééquilibré pour Malware:", len(malware_resampled))
print("Taille de l'ensemble de données rééquilibré pour Phishing:", len(phishing_resampled))
print("Taille de l'ensemble de données rééquilibré pour Spam:", len(spam_resampled))
```

```python
#Trop Gourmand sur mon pc


from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from sklearn.preprocessing import OneHotEncoder

# Concaténation des ensembles de données
all_data = pd.concat([benin_data, malware_data, phishing_data, spam_data], ignore_index=True)

# Encodage one-hot des variables catégorielles
categorical_columns = all_data.select_dtypes(include=['object']).columns
all_data_encoded = pd.get_dummies(all_data, columns=categorical_columns)

# Séparation des fonctionnalités et de la cible
X = all_data_encoded.drop('Class', axis=1)
y = all_data_encoded['Class']

# Encodage des variables catégorielles en nombres
encoder = OneHotEncoder()
X_encoded = encoder.fit_transform(X)

# Définir le nombre de données à sous-échantillonner pour chaque classe
undersample_size = min(len(benin_data), len(malware_data), len(phishing_data), len(spam_data))

# Sous-échantillonnage aléatoire
undersampler = RandomUnderSampler(sampling_strategy={'Benin': undersample_size,
                                                     'Malware': undersample_size,
                                                     'Phishing': undersample_size,
                                                     'Spam': undersample_size})
X_resampled, y_resampled = undersampler.fit_resample(X_encoded, y)

# Sur-échantillonnage (SMOTE) des classes minoritaires
smote = SMOTE(sampling_strategy='auto')  # 'auto' pour équilibrer toutes les classes
X_resampled, y_resampled = smote.fit_resample(X_resampled, y_resampled)

# Recréer un DataFrame avec les données rééchantillonnées après SMOTE
all_resampled = pd.DataFrame(data=X_resampled, columns=encoder.get_feature_names_out(X.columns))
all_resampled['Class'] = y_resampled

# Afficher le décompte des classes après rééchantillonnage
print(all_resampled['Class'].value_counts())
```
